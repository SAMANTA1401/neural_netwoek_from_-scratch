DENSE LAYER CLASS

```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data
nnfs.init()
```

```python
# starting next layer of input layer  belongs to hidden layer

# Dense layer
class Layer_Dense:
 # Layer initialization
 def __init__(self, n_inputs, n_neurons):
 # Initialize weights and biases
   self.weights = 0.01 * np.random.randn(n_inputs, n_neurons) # doesnt need transpose of weight
   self.biases = np.zeros((1, n_neurons))

 # Forward pass
 def forward(self, inputs):
 # Calculate output values from inputs, weights and biases
   self.output = np.dot(inputs, self.weights) + self.biases
```

```python
class Layer_Dense2:
 # Layer initialization
 def __init__(self, n_inputs, n_neurons):
 # Initialize weights and biases
   self.weights = 0.01 * np.random.randn(n_neurons, n_inputs) # need transpose of weights
   self.biases = np.zeros((1, n_neurons))

 # Forward pass
 def forward(self, inputs):
 # Calculate output values from inputs, weights and biases
   self.output = np.dot(inputs, np.array(self.weights).T) + self.biases
```

```python
# np.random.randn(n_inputs, n_neurons) generates a matrix of random numbers from a standard normal distribution.
# Parameters:
# n_inputs: Number of rows (input features)
# n_neurons: Number of columns (neurons in the layer)
# Return Value:
# A 2D NumPy array (matrix) with shape (n_inputs, n_neurons)
# Standard Normal Distribution:
# Mean: 0
# Standard Deviation: 1
# Range: (-∞, ∞)
```

```python
D1 = Layer_Dense(4,3)
```
